## Get information through Wikipedia Website on TV stations' article.

### For this purpose, I made a recursive function using BeautifulSoup python module

```python
#Get information from TV Station page on Wikipedia
page = requests.get('https://pt.wikipedia.org/wiki/Categoria:Canais_de_televis%C3%A3o_do_Brasil_por_rede')
```

```python
#Visualization of TV Station page and html tags
soup = BeautifulSoup(page.text, 'html.parser')
print(soup.prettify())
```

```python
#Testing the "soup.find" function
soup.find(id='firstHeading').string.split(':')[0] == 'Categoria'
```

```python
#Development of the scrap_broad recursive function
#The function access the links containing TV Networks and TV Stations
#If the page is a TV Station the function calls itself recursively into new page until a wrong page be accessed

page = 'https://pt.wikipedia.org/wiki/Categoria:Canais_de_televis%C3%A3o_do_Brasil_por_rede'


def scrap_broad(page, broadcasters = [], first_page = True):
  
    '''Recursive Function for get all TV Stations
    
       Input: page - page to scrap information
              broadcasters - list to add new TV broadcasters
              first_page - just to check if is the initial page
              
       Output: broadcasters - list containing all TV Stations
    '''
    
    
    #Get the informed page
    page = requests.get(page)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    #If the initial page informed the script will work on "div" tag and "mw-subcategories" id
    if(first_page):
        
        #Try to get a link from TV Network and open the scrap_broad function recursively
        #sending the updated list "broadcasters" to the function
        try:
            subcat_lnk = soup.find('div',id='mw-subcategories').findAll('a')
            for broad_lnk in subcat_lnk:
                print('\n')
                print(broad_lnk['title'])
                print('\n')
                broadcasters = scrap_broad('https://pt.wikipedia.org'+broad_lnk['href'], broadcasters, False)

        except:
            subcat_lnk = None
        


    #If it isn't the initial page the script will work on "div" tag and "firstHeading" and "mw-pages" id      
    else:                                 
        title = soup.find(id='firstHeading').string.split(':')[0]
        
        #If title is "Categoria" means that the page is a TV station and must to be analyzed
        if title == 'Categoria':
            
            #Try to get the name of TV Station link and access the link recursively to get new information
            #sending the updated list "broadcasters" to the function
            try:
                page_lnk = soup.find('div',id="mw-pages").findAll('a')
                
                for broad_lnk in page_lnk:

                    broad_name = broad_lnk['title']
                    print(broad_name)
                    broadcasters.append(broad_name)
                    broadcasters = scrap_broad('https://pt.wikipedia.org'+broad_lnk['href'], broadcasters, False)

            except:
                page_lnk = None
            
            
    return broadcasters
```

```python
#Call the function on the first page and print result
tvStation = scrap_broad('https://pt.wikipedia.org/wiki/Categoria:Canais_de_televis%C3%A3o_do_Brasil_por_rede')
```

## Take all wrong Stations off the list.

```python
#Take off all wrong names in the "words_excluded"
words_excluded = ["Predefinição","Usuário","Lista","Discussão"]
  
print('TV Brands excluded: \n')

#In order to loop and delete at same time was used "reversed" function to reverse the list and keep the order
for tv in reversed(tvStation):
  for word in words_excluded:
    if(tv.find(word) != -1):
      print(tv)
      tvStation.remove(tv)
      break
```

## Download Logo Images

I used the google_images_download to search on Google Images the name of each TV Stations and Download their Logos

```python
#The module "google_images_download" needs the chromedriver.exe in order to work properly
driverpath = r'chromedriver.exe'
from google_images_download import google_images_download

def images_down(keyws,limit = 5,chromedriver = driverpath,offset = None,color_type = None,size = None,usage_rights = None):
    
    '''Function to Download the Logos of all TV Stations
       
       Input: keyws - keywords (TV Stations) to search on google images
              limit - The number of images that will be downloaded
              chromedriver - necessary module to function
              ** The others arguments is just for futher test
    '''
    
    
    arguments = {
            "keywords" : keyws,
            "limit" :limit,
            "chromedriver":chromedriver,
            "offset" : offset,
            "color_type" : color_type,
            "size" : size,
            "usage_rights" : usage_rights
            }
    
    response  = google_images_download.googleimagesdownload()
    
    #Try to get the images requested on "keyws"
    try:
      response.download(arguments)
      
    #Due the different code of brazilian words, sometimes this Unicode error happens and they are just ignored
    except UnicodeDecodeError: 
      pass
    
```

```python
#The function "images_down" is called on all TV Station list
#Download the 5 first results from google images of each TV Station
for tv in tvStation:
  images_down("Logo "+tv)
  print()
```
